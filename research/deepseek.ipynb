```python
file_path = "research/real_data_sentiment_analysis_results.json"

# Load reviews from JSON file
try:
    with open(file_path, "r", encoding="utf-8") as file:
        reviews = json.load(file)
except FileNotFoundError:
    print(f"Error: File not found at {file_path}. Check the path and try again.")
    exit()

# Construct prompt for DeepSeek AI
prompt_text = "Analyze the following restaurant reviews and provide insights on common complaints, suggestions for improvement, and recurring themes:\n\n"

for review in reviews:
    prompt_text += f"Review: {review['review']}\nSentiment: {review['sentiment_description']}\n\n"

# Send request to DeepSeek AI via OpenRouter
completion = client.chat.completions.create(
    extra_body={},
    model="deepseek/deepseek-r1:free",  # Use DeepSeek's free model
    messages=[
        {"role": "system", "content": "You are a restaurant owner wanting to understand insights on customer feedback."},
        {"role": "user", "content": prompt_text},
    ]
)

# Print AI-generated insights
print("DeepSeek AI Insights:", completion.choices[0].message.content)
```
